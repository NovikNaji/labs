```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
import numpy as np
from skimage import draw
import torch
from PIL import Image    #  pip install Pillow


def plot2d(data):
    plt.imshow(data)
    plt.show()

def plot3d(data):
    fig = plt.figure(figsize=(100, 100))
    grid = ImageGrid(fig, 111, nrows_ncols=(1, 3), axes_pad=0.02)
    for ax, im in zip(grid, data):
        ax.imshow(im)
    plt.show()


with Image.open('/content/drive/MyDrive/лабы/ml/kot.jpg') as img:
  img.load()
img = np.array(img.convert("L"))



plt.imshow(img)
plt.show()

input_data = np.expand_dims(img, axis=0)

print(input_data.shape)
```

```python
from torch import nn

conv1 = nn.Conv2d(in_channels=1, out_channels=1, kernel_size=(13, 13), stride=1)

input_tensor = torch.from_numpy(input_data).to(torch.float32)
print(input_tensor.size())

output_tensor = conv1(input_tensor)
print(output_tensor.size())

np_out = output_tensor.detach().numpy()
print(np_out.shape)

plot3d(np_out)
```
